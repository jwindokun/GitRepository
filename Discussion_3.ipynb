## Introduction

Albanese et al., in their article give a concise introduction to the field of Recommender Systems, discuss the historical context, and the statistical, mathematical and social background of the different implementations of recommenders and they discuss the limitations of such systems such as “cold start”, “overspecialization”, SVD with latent factors and so forth.

    Search is what you do when you are looking for something while discovery is when something wonderful that you did not know existed, or didn’t know how to ask for, finds you.
    
Ideally, therefore, we would want our Recommender System to do this, amaze the user with something new that they did not expect to find/see.


# Types of Recommenders

Recommenders can be divided into two major divisions:
1.	Content based – that looks into the attributes of items or users and finds similarities between them and uses this to recommend/predict other items

2.	Collaborative based – uses items or users and groups them together based on similarity and based on this recommends/predicts other items that the user/items have in common
Hybrid Solutions

Hybrid Solutions use other measures or a combination of Content and Collaborative (one after the other) to improve on the efficacy metrics of the recommender. SVD, PCA, and Latent Analysis aim to find a subset of user/item attributes that might explain user/item preferences and use this to recommend/predict items 
Recommenders as a Social Choice Problem – Albanese et al’s Model.


# Recommenders as a Social Choice Problem – Albanese et al’s Model.

The authors of this paper look at a recommendation as a social choice problem and invoke the PageRank Algorithm that makes Google the powerhouse it is today as a search engine.
Concisely they combine the user recommendations, which is specific to the user with the global rankings of all the items by all the users.
In a nutshell, if globally most users that buy item A also buy item B, then when ranking items for user 1, they not only consider the ranked items, they also consider the globally ranked items, therefore even if item B does not appear in the personalized ranked items for user 1, due to the fact that it is highly ranked globally for all users, it will be added to the personalized list for user 1.
Albanese et al., describe their method as:

    Computes customized recommendations by combining – in a novel way – intrinsic features of multimedia objects (low level and semantic similarities), past behaviour of individual users and overall behavior of the entire community of users 
    
Practically, they weigh the user matrix of recommendations by alpha and the collective (population) matrix of recommendations by (1-alpha) and gradually increase the level of alpha from 0, no information about the user at all to 0.5. Similarly they vary the parameters gamma and delta which account for the fact that the user will select one of the population recommendations items and that the user will randomly select something unconnected based on the experience with the user.    

# Practical Use of this Model

Netflix for example recommends movies based “Trending Now” which are basically movies that the majority of users are watching now, “Popular on Netflix” – popular movies over a given period of time and “Top Picks”, which are recommendations that Netflix has made for me based on my viewing habits. 
With the use of the Social Choice Model, Netflix could replace these 3 different recommenders with a single recommender that takes into account my tastes and preferences, the top picks and trending now movies.
While this might initially seem to limit my choices, in actual fact by me providing feedback on which of these movies I actually like to see (by clicking on them or watching them) Netflix will improve on its recommendation engine (increasing the value of gamma in the equation above).
This is of particular importance where the amount of space the available to show the choices are limited (think smartphone screen versus a computer monitor or TV)


# Conclusion

Albanese et al., have come up with a Social Choice Model, which incorporates the global community rating in recommendations. This is similar to the PageRank Algorithm that Google uses to deliver its search results which has been found to be most effective. 
This model is particularly important where the ability to display choices are limited and most importantly it allows the user the ability to “discover … something wonderful”
