## Summary

### Recommendations should evolve over time - Charlin et. al., (2015)

The talk is based on a paper titled by Dynamic Poisson Factorization by Charlin et. al.(2015) and is on recommender systems.
The authors state that traditionally, recommender systems use latent factors (user observations or variables) to explain user preferences for items. And similarly, items are usually regarded as having static properties, and therefore the whole dataset of users, user properties, preferences, items and item properties are all regarded as static and unchanging with time.

However, this is not true as both users and items evolve over time. Users for example grow older, their tastes vary with time, and similarly, items vary over time, their prices, attributes and so forth, and even availabilty vary over time. All this is not captured in the traditional recommender system, and therefore one can assume that as time goes by traditional recommender systems will make recommendations that are more and more out of date with what the preferences of the user actually are.

The authors state that this becomes more of an issue when recommender systems have users that have been with them for years (Amazon for example or ArXiv.org or even Netflix), over the years one would assume that the users change with time and therefore it is of utmost importance that the recommendation system takes into account and understands this.

The Dynamic Poisson Factorization Model (dFP) that the authors create takes this into account by taking the latent factors and instead of keeping them static over time, allow them to vary over epochs of time. 

dFP specifically is a probalistic model that uses the Poisson distribution and Gamma distribution to model implicit  user inputs (time series of clicks). The Poisson distribution is used to model user clicks while the Gaussian distribution is used to model users/item factors.

## Implementation ideals for dFP in Recommender Systems

One of the advantages of dFP is that it is time aware. Almost all recommender systems that have users that have been with them for some period of time would benefit from this as both user preferences and items change with time.
For this excercise I would use Netflix.
Netflix's recommendation engine is presently built on what users play, search or rate (http://www.wired.com/2013/08/qq_netflix-algorithm/) and is based on content similarity (movies) and collaboration (similar users like you to acertain your preference). Netflix also tries to implment a context based (time of day) recommender which will try and recommend different movies based on the time of the day.

However, Netflix does not take into consideration that its users change with time, and that the movies they might have seen in the past may no longer be relevant to them. For example, I have been a Netflix customer since 2008 (8 years ago)and I still get recommendations for genres for movies that I will never see again. 

Netflix should revamp their recommendation system to give lesser weights to movies that I and other users have seen in the past (collaborative filtering) and therefore the recommendations that it gives will be more relevant. 

Also in terms of movie types and shows, in the past decade or so "Reality TV" has become quite big, Netflix's recommendation system should keep this in mind and in its recommendations, shows that belong to the genre "Reality TV" should probably be scored higher.
